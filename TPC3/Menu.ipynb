```python
import random
def menu():
    print('''Escolha as seguintes opções do Menu:
(1) Criar uma lista aleatória 
(2) Ler uma lista
(3) Soma dos elementos da lista
(4) Média dos elementos da lista
(5) Maior elemento da lista
(6) Menor elemento da lista
(7) Verificar se a lista está ordenada por ordem crescente
(8) Verificar se a lista está ordenada por ordem decrescente
(9) Procurar um elemento na lista
(0) Sair''')
    return


def criarLista(tamanho):
    listaRand=[]
    n = 0
    while n< tamanho:
        elemRand= random.randint(1,100)
        listaRand.append(elemRand)
        n = n + 1
    return listaRand

def lerLista(tamanho):
    lista = []
    n= 0
    while n< tamanho:
        elem = int(input("Introduza o "+str(n+1)+" elemento da lista:"))
        lista.append(elem)
        n = n + 1
    return lista

def somaLista(lista):
    soma= 0
    for n in lista:
        soma = soma + 1
    return soma

def mediaLista(lista):
    media = somaLista/len(lista)
    return media

def maiorElemLista(lista):
    maior = lista[0]
    for n in lista:
        if n > maior:
            n = maior
    return maior

def menorElemLista(lista):
    menor = lista[0]
    for n in lista:
        if n < menor:
            n = menor
    return menor

def ordenadaCrescente(lista):
    resultado= True
    n=0
    while n+1<len(lista):
        if lista[n]>lista[n+1]:
            resultado=False
        n=n+1
    return resultado

def ordenadaDecrescente(lista):
    resultado= True
    n=0
    while n+1<len(lista):
        if lista[n]<lista[n+1]:
            resultado=False
        n=n+1
    return resultado

def elemLista(lista, elem):
    posiçaoLista= []
    for n in range(len(lista)):
        if lista[n]==elem:
            posiçaoLista.append(n+1)
    if len(posiçaoLista)==0:
        posiçaoLista.append(-1)
    return posiçaoLista

menu()
function=int(input("Qual operação do menu deseja realizar?"))
listaRes=[]

while function!=0:
    if function==1:
        listaRes.clear()
        listaRes=criarLista(int(input("Quantos elementos deseja na sua lista?")))
    if function==2:
        listaRes.clear()
        listaRes=criarLista(int(input("Quantos elementos gostaria de por na lista?")))
    if function==3:
        print("")
        print("A soma dos elementos da sua lista é:",somaLista(listaRes))
    if function==4:
        print("")
        print("A média dos elementos da sua lista é:",mediaLista(listaRes))
    if function==5:
        print("")
        print("O maior elemento da sua lista é:",maiorElemLista(listaRes))
    if function==6:
        print("")
        print("O menor elemento da sua lista é:",menorElemLista(listaRes))
    if function==7:
        if ordenadaCrescente(listaRes)==True:
            print("")
            print("Sim.")
        else:
            print("")
            print("Não.")
    if function==8:
        if ordenadaDecrescente(listaRes)==True:
            print("")
            print("Sim.")
        else:
            print("")
            print("Não.")
    if function==9:
        posLista=elemLista(listaRes,int(input("Qual o elemento que procura?")))
        if posLista[0]==-1:
            print("")
            print("O elemento que procura não se encontra na lista.")
        else:
            print("")
            print("O elemento que procura encontra-se na(s)",posLista,"posição(ões).")
function = int(input("O que pretende fazer?"))
print("")
print("A sua lista é:", listaRes)











```
